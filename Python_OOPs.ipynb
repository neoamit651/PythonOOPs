1. What is Object-Oriented Programming (OOP)?
  - Object-oriented programming (OOP) in Python is a paradigm for structuring programs by bundling related properties and behaviors into individual objects. This approach organizes code into logical, reusable components called classes, which serve as blueprints for creating specific objects.
  - Key concepts of OOP in Python
The four core principles of OOP in Python are encapsulation, inheritance, abstraction, and polymorphism
    - Encapsulation: Encapsulation groups data (attributes) and the methods that operate on that data within a single class. This practice helps protect sensitive data and prevent unauthorized changes. Python uses naming conventions to signify public, protected (using a single underscore prefix like _attribute), and private members (using a double underscore prefix like __attribute to trigger name mangling).
    - Inheritance: Inheritance enables a new class (subclass) to inherit attributes and methods from an existing class (superclass), promoting code reuse and establishing class hierarchies. Python supports various types of inheritance, including single, multiple, multilevel, and hierarchical. The super() function is commonly used to access methods from the parent class within a child class.
    - Abstraction: Abstraction focuses on presenting essential features while concealing complex implementation details, simplifying interaction and improving program efficiency. In Python, the abc module is used for abstraction, allowing the creation of abstract classes (which cannot be instantiated) and abstract methods (which must be implemented by subclasses).
    - Polymorphism: Polymorphism allows a single function, operator, or method to exhibit different behaviors depending on the object type it is used with. This can be achieved through method overriding, where a subclass provides its own implementation of a parent class method, or function polymorphism, as seen with functions like len() which work differently for various object types like strings or lists.
- Python is an object-oriented language, and almost every element—including strings, lists, integers, and functions—is an object with its own methods and attributes.

  

2. What is a class in OOP?
  - In object-oriented programming (OOP) in Python, a class is a blueprint or template for creating objects. A class bundles together data and the functions that operate on that data into a single, reusable unit. It defines the properties (attributes) and behaviors (methods) that all objects created from it will have.
  - Key components of a Python class
    - Attributes: Variables that hold data associated with a class. There are two types:
      - Class attributes: Belong to the class itself and are shared by all instances of the class. For example, all Dog objects might share a species attribute with the value "canine".
      - Instance attributes: Variables specific to an object. Each instance has its own copy. In a Dog class, name and age would be instance attributes, as each dog has a unique name and age.
    - Methods: Functions defined inside a class that define the behaviors of objects created from that class. A method's first parameter is always self, which refers to the specific instance the method is being called on.

    - Everything is an object: In Python, everything is fundamentally an object, including numbers, strings, lists, functions, and even classes themselves.
  Creating an object
  - We create an object by "instantiating" a class. This calls the class like a function, and Python's __init__ method is automatically called to set up the object's initial state.

7. What is polymorphism in OOP?
  - In Python, polymorphism is the ability of an object to take on many forms, allowing a single interface to be used for different underlying data types or classes. The name comes from the Greek words poly ("many") and morph ("form").
  - This concept enables we to write code that is flexible, reusable, and more organized. Instead of needing different functions for every object type, we can use a single function or method that behaves differently depending on the object it is called on.
  - Key ways polymorphism is implemented in Python
    - 1. Method overriding (runtime polymorphism): Method overriding is the most common form of polymorphism in Python. It occurs when a subclass provides its own specific implementation of a method that is already defined in its parent class. When we call this method on an object, Python determines which version to execute at runtime, based on the object's actual class.
    - 2. Duck typing: Python is a dynamically typed language and relies on the "duck typing" principle. This means that an object's suitability is determined by its behavior (its methods) rather than its explicit class. The phrase "if it walks like a duck and quacks like a duck, then it must be a duck" perfectly illustrates this concept.
    - 3. Operator overloading: Polymorphism is also seen with built-in operators. The same operator can behave differently depending on the data types it is applied to.
    - 4. Function polymorphism: Some of Python's built-in functions, like len(), are polymorphic. They can work with multiple data types and produce a result that is specific to that type.
  - Benefits of polymorphism
    - Code reusability: Write generic code that can work with objects of many different types, reducing code duplication.
    - Easier maintenance: When we add a new class, we don't need to change the existing code that works with it, as long as the new class conforms to the same interface.
    - Improved readability: Avoid complex if/else statements that check for object types, making the code cleaner and easier to understand.
    - Loose coupling: Code becomes more modular because objects interact based on a common interface, not on specific implementation details.


8. How is encapsulation achieved in Python?
  - Encapsulation is achieved in Python primarily through naming conventions and, for private members, a mechanism called name mangling. Unlike other object-oriented languages that have explicit keywords like private and protected, Python's approach is more of a convention that programmers are expected to respect. The core of encapsulation—bundling data and methods within a class to control access—is still a fundamental part of Python's object-oriented paradigm.
  - Python achieves different levels of encapsulation using the following access modifiers:
  - 1. Public members: By default, all attributes and methods in a Python class are public. This means they can be accessed from anywhere, both inside and outside the class.
  - 2. Protected members: Protected members are intended for internal use within the class and its subclasses. We can indicate a protected member by prefixing its name with a single underscore (_).
  - 3. Private members: Private members are only accessible within the class itself. To make an attribute or method private, we prefix its name with a double underscore (__). When we do this, Python's interpreter renames the attribute through a process called "name mangling." This makes it more difficult (but not impossible) to access from outside the class.
  - 4. Getter and setter methods: To ensure proper encapsulation, it is common to use public "getter" and "setter" methods to access or modify protected or private data. This gives we control over how data is accessed and allows for validation.

12.  What is method overriding in OOP?
  - In Python, method overriding is an object-oriented programming (OOP) concept that allows a subclass (or child class) to provide a specific implementation for a method that is already defined in one of its superclasses (or parent classes). This enables a child class to change or extend the behavior of an inherited method to suit its own needs.
  - How it works
    - Method overriding relies on class inheritance. When we create a method in a child class with the exact same name and number of parameters as a method in the parent class, the child's method replaces the parent's version.
    - Key characteristics
      - Requires inheritance: Method overriding only occurs when there is a parent-child relationship between classes.
      - Same method signature: The method in the child class must have the same name and the same number of parameters as the method in the parent class.
      - Polymorphism: Overriding is a key feature of runtime polymorphism. The method that is called depends on the type of object, not the type of the variable referencing the object. This allows objects of different classes to respond differently to the same method call.
    - Extending, not just replacing, the parent method
    - Often, we don't want to completely replace the parent method's behavior. Instead, we may want to execute the original logic and then add or modify it. We can do this by using the super() function to call the parent's method from within the overridden child method.





13. What is a property decorator in Python?
  - A property decorator in Python is a built-in feature that allows we to define methods in a class that can be accessed like attributes. It is the Pythonic way to implement getters and setters, which provides controlled access to an object's data.
  - How the @property decorator works
  - The @property decorator is used with three corresponding methods to define how an attribute is handled:
    - Getter: The method decorated with @property retrieves the value of the attribute.
    - Setter: A method decorated with @<property_name>.setter sets the value of the attribute. This is where we can add validation logic.
    - Deleter: A method decorated with @<property_name>.deleter deletes the attribute from the object.
  - By using this approach, we can change the internal implementation of a class without altering the public interface. If we initially define a simple public attribute and later decide it needs validation, we can convert it to a property without breaking code that already uses our class.
  - Advantages of using @property
    - Encapsulation: It allows we to protect the internal state of an object and manage how its attributes are accessed and modified, enforcing object-oriented principles.
    - Data Validation: we can add custom validation logic to the setter method to ensure an attribute is always set to a valid value.
    - Computed Attributes: The getter can dynamically calculate and return a value based on other attributes, so the value doesn't need to be stored directly.
    - Improved Readability: Using the @property decorator allows our code to treat methods like simple attributes, which results in cleaner and more intuitive syntax.
    - Refactoring without breaking client code: If we need to add logic to a public attribute later, we can convert it into a property without changing the way other parts of our program interact with it.




14. Why is polymorphism important in OOP?
  - Polymorphism is important in object-oriented programming (OOP) because it allows for code reusability, flexibility, and extensibility. This core concept enables objects of different classes to be treated as objects of a common superclass or interface, providing a single, consistent way to interact with them.
  - Key reasons why polymorphism is so important:
  - It enhances code reusability
    - By relying on a single, shared interface for related classes, polymorphism reduces the need for redundant code.
    - One interface, many implementations: We can write one general method that works with objects of different types, rather than writing a separate method for each type.
    - Example: A draw() method: A general draw() function can be called on any object that inherits from a Shape class (e.g., Circle, Square, Triangle). The correct draw() implementation is executed automatically for each specific shape object.
  - It improves flexibility and extensibility
    - Polymorphism allows we to build systems that are easily extended with new functionality without changing existing code.
      - Facilitates "plugin" architecture: We can add new subclasses that implement a shared interface and have our existing code work with them immediately. This adheres to the "Open-Closed Principle," which states that software entities should be open for extension but closed for modification.
      - Example: A Payment system: A payment processing module can be written to accept a generic Payment object. If a new payment method, such as a cryptocurrency, is introduced, we simply create a new subclass that implements the Payment interface. The core processing module does not need to be updated.
   - It simplifies code and makes it more maintainable
      - Using polymorphism helps reduce the need for complicated, error-prone conditional logic and switch statements.
        - Centralized logic: We can define a method in a parent class and then have all subclasses provide their own specific implementation. To update the logic, we only need to modify the code in one place or the specific subclass, which simplifies maintenance.
        - Cleaner code: Code that uses polymorphic methods is more readable and easier to follow, as it focuses on the object's behavior rather than its specific type. This leads to a better overall structure and design.
      - It enables abstract design
        - Polymorphism encourages programming to an interface rather than a specific implementation. This promotes decoupling between components.
          - High-level focus: By interacting with objects through a common interface or abstract class, developers can think at a higher level of abstraction, focusing on what an object does rather than how it does it.
          - Example: An Animal shelter: A program can manage a collection of Animal objects, where each can be a Dog, Cat, or Bird. The program doesn't need to check the specific type of each animal, as the correct overridden method will be called automatically at runtime.
      - Two main types of polymorphism
        - Compile-time polymorphism (Method Overloading): Multiple methods within the same class share the same name but have different parameter lists. The compiler determines which method to call at compile-time based on the arguments provided.
        - Runtime polymorphism (Method Overriding): A subclass provides its own specific implementation of a method that is already defined in its parent class. Which version of the method to execute is determined dynamically at runtime based on the actual object type.




15. What is an abstract class in Python?

They cannot be instantiated directly and often contain one or more abstract methods, which are declared but lack implementation. This enforces a consistent structure across a family of related classes.
    - Key characteristics
      - Cannot be instantiated: We cannot create an object from an abstract class directly. Attempting to do so will result in a TypeError.
      - Uses the abc module: To create an abstract class, we must inherit from the ABC class and use the @abstractmethod decorator for its abstract methods.
      - Enforces implementation: Subclasses are required to provide a concrete implementation for all abstract methods inherited from the abstract base class.
      - Can have concrete methods: Abstract classes can also contain regular, implemented methods that are inherited by their subclasses, promoting code reuse.
  - When to use abstract classes
    - Defining a standard interface: Use an abstract class to define a standard set of methods that all subclasses must have. This is useful for building plugin systems or different drivers for the same functionality.
    - Promoting consistency: Abstract classes ensure that all related classes have a similar structure, making the codebase more organized and easier to maintain.
    - Facilitating polymorphism: We can write code that can operate with objects of different types as long as they conform to the same abstract base class. This allows we to treat different subclasses in a uniform way.

18. What is multiple inheritance in Python?
    - Multiple inheritance is a feature in Python where a class can inherit attributes and methods from more than one parent class. This allows the child class to combine functionalities from all its parent classes, promoting code reuse.
    - How it works- To implement multiple inheritance, we list all the parent classes inside the parentheses of the child class's definition:
    - The diamond problem and Method Resolution Order (MRO)
      - A potential complexity with multiple inheritance is the "diamond problem," which occurs when two parent classes inherit from a common grandparent class. If a method with the same name exists in multiple parent classes, it creates an ambiguity regarding which one the child class should inherit.
          - Python resolves this conflict using the Method Resolution Order (MRO), which defines the order in which base classes are searched for a method or attribute.
            - The MRO is determined using the C3 linearization algorithm, which ensures a consistent and predictable search order.
            - The search is performed in a depth-first, left-to-right manner. The first class that contains the requested method or attribute is used.
            - We can view the MRO of any class by calling its mro() method or by accessing the __ mro __ attribute.
      - The super() function
        - When dealing with multiple inheritance, the super() function is used to call methods from a parent class according to the MRO. This is particularly useful for cooperating with parent classes and avoiding manual, explicit calls to parent methods, which can lead to the "diamond problem" bug where the same method is called multiple times.

20. What is the significance of the ‘super()’ function in Python?
  - The super() function in Python provides a way for a subclass to call a method that is defined in one of its parent or "superclass" classes. Its primary significance is enabling cooperative inheritance, which allows for robust and maintainable class hierarchies, especially with multiple inheritance.
  - Key roles of super()
    - Calling parent constructors (): The most common use case is within a subclass's __ init __ method. By calling super().__ init __(), we ensure that the parent class's attributes and methods are correctly initialized before the child class adds its own specific attributes. This prevents we from having to manually repeat initialization code.
    - Method overriding and extension: super() allows we to override a parent's method in a subclass while still being able to call the original parent method. This lets we add new or modify existing functionality without completely replacing the inherited behavior.
    - Enabling cooperative multiple inheritance: super() is essential for managing complex class hierarchies, especially those with multiple inheritance. It uses a consistent and predictable Method Resolution Order (MRO) to determine which parent class's method to call next. This prevents the "diamond problem," where a method from a common ancestor is mistakenly called multiple times.
    - Avoiding hardcoded class names: With super(), we do not need to explicitly name the parent class. This makes our code more flexible. If we change the parent class, super() will automatically adapt, so we don't have to find and replace the old class name throughout our code.
    - Promoting code reusability and maintainability: By leveraging parent class functionality instead of rewriting it, super() makes code cleaner and more efficient. It centralizes shared logic in parent classes, which simplifies maintenance and updates.
  - How super() works with Method Resolution Order (MRO)
    - The Method Resolution Order is the order in which Python searches for inherited methods. We can inspect the MRO of any class by calling its __ mro __ attribute or the mro() method.
    - For example, given a diamond-shaped inheritance hierarchy where D inherits from B and C, which both inherit from a common ancestor A, the MRO ensures that super() calls the methods in the correct, non-repetitive order:
      - D first looks for the method in its own class.
      - If not found, it checks its first parent, B.
      - Then, it checks B's first parent, A.
      - Finally, it checks C, before resorting to the base object class.
    - This predictable order is determined by the C3 linearization algorithm, which avoids the inconsistencies that plagued older Python versions.



21. What is the significance of the __del__ method in Python?
  - In Python, there is a key distinction between the del keyword and the __del__() method. The keyword is a general-purpose statement for unbinding names, while the method is a class destructor used for cleaning up resources.
  - del statement: Removing a reference
    - The del statement's significance is that it removes a name (a variable, attribute, or item) from its scope. It does not directly cause an object to be removed from memory. Instead, it deletes a reference to that object. Once the reference count for an object drops to zero (meaning nothing refers to it anymore), Python's garbage collector can reclaim the memory.
      - Primary uses of the del statement
        - Deleting variables: Unbinds a variable name from the object it references.
        - Deleting mutable collection items: Removes items from lists, dictionary keys, or other mutable objects.
        - Deleting slices: Removes multiple elements from a list at once.
    - __ del __() method: Custom destructor
      - The __ del __() method is a special "magic" or "dunder" method that can be defined inside a class. It is significant because it provides a mechanism for an object to perform custom cleanup just before it is destroyed by the garbage collector.
      - Significance and limitations of __ del __()
        - Resource cleanup: It is designed for tidying up external resources, such as closing file handles, network connections, or database sessions, that are associated with an object.
        - Not a guarantee: The timing of when __ del __() is called is not guaranteed. Python's garbage collector determines when an object is destroyed, so the destructor may not run immediately after the last reference is removed. It may not even run at all if the program exits abnormally.
        - Avoid for critical tasks: Due to its unreliable timing, __ del __() should not be used for critical resource management. The recommended practice for reliable cleanup is to use a with statement and a context manager, which handles resources deterministically.
        - Ignoring exceptions: Any exceptions that occur within the __ del __() method are ignored by Python, which can lead to silent failures that are difficult to debug.

24.  What is method chaining in Python OOP?
     - In Python's object-oriented programming (OOP), method chaining is a technique that allows we to call multiple methods on a single object in a continuous line of code. This creates a more readable and concise flow of operations, often referred to as a "fluent interface".
     - The core principle is that each method in the chain performs an action and then returns the object itself (self), which can be immediately used to call the next method. This eliminates the need for intermediate variables to store the object at each step.
     - How method chaining works
     - To make a method chainable, we must explicitly return the current instance of the class (self) at the end of the method's definition.
     - Benefits of method chaining
        - Concise code: It reduces the number of lines and temporary variables needed to perform a sequence of operations.
        - Readability: It can make code easier to read and understand by expressing a clear, sequential flow of actions.
        - Fluent APIs: It is essential for creating elegant and expressive interfaces, often used in libraries for tasks like building queries or transforming data.
        - Clarity: It clearly shows that multiple operations are being performed on the same object.
     - Drawbacks of method chaining
        - Difficult debugging: If an error occurs in the middle of a long chain, it can be harder to isolate the exact method that caused the problem.
        - Reduced clarity: Overly long or complex chains can become difficult to read and understand, especially for developers unfamiliar with the code.
        - Risk of side effects: If a method unexpectedly returns a different object instead of self, it can break the chain and lead to errors.
        - Performance overhead: In the case of immutable objects (like Python's built-in strings), each chained method call creates and returns a new object, which can consume more memory.
      - Real-world examples
        - Method chaining is a common pattern in many popular Python libraries:
          - Pandas: DataFrames often use method chaining for data manipulation tasks, such as filtering, sorting, and aggregating data.
          - SQLAlchemy: The library for building SQL queries uses method chaining to construct complex database queries in a readable way.
          - String methods: Many built-in string methods, such as strip(), lower(), and replace(), can be chained because each returns a new string object.




25. What is the purpose of the __ call __ method in Python?
    - The __ call __ method in Python allows an instance of a class to be called as if it were a function. This makes the object "callable," which is useful for creating flexible, reusable code that can maintain state between calls.
    - How __ call __ works: When we define the special __ call __ method within a class, any instance of that class can be executed using the function-call syntax (e.g., object()). The code inside the __ call __ method is what runs when the instance is called.
    - Common use cases for __ call __
      - 1. Creating "stateful" objects: A callable object can maintain and modify its internal state between calls. This allows it to remember information from previous interactions.
      -  2. Implementing class-based decorators: A decorator modifies the behavior of a function or method. A callable class provides a more robust way to build complex decorators than using a simple function. It can accept arguments at both the decoration stage (when the decorator is applied) and the function-call stage (when the decorated function is executed).
      - 3. Designing clean and convenient APIs: The __ call __ method can make an object's interface more intuitive by allowing a single, primary action to be invoked directly on the object.
      - 4. Creating function factories: We can create objects that act as factories for other functions. The __init__ method can set up the configuration, while __call__ generates a new function based on that setup.
    - __ call __ vs. __ init __
    - It is important not to confuse __ call __ with __ init __.
      - The __ init __ method is the constructor. It is executed once when an object is created to initialize its state.
      - The __ call __ method is an instance method that is executed every time the object is "called." It can be invoked multiple times throughout an object's lifetime.

```python
# 2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.
# Ans.
import math
from abc import ABC, abstractmethod

# Abstract class `Shape` inheriting from `ABC`
class Shape(ABC):
    @abstractmethod
    def area(self):
        """Abstract method to calculate the area of the shape."""
        pass

# Derived class `Circle`
class Circle(Shape):
    def __init__(self, radius):
        """Initializes a Circle with a radius."""
        if radius < 0:
            raise ValueError("Radius cannot be negative.")
        self.radius = radius

    def area(self):
        """Calculates the area of the circle using the formula pi * r^2."""
        return math.pi * (self.radius ** 2)

# Derived class `Rectangle`
class Rectangle(Shape):
    def __init__(self, width, height):
        """Initializes a Rectangle with a width and height."""
        if width < 0 or height < 0:
            raise ValueError("Width and height cannot be negative.")
        self.width = width
        self.height = height

    def area(self):
        """Calculates the area of the rectangle using the formula width * height."""
        return self.width * self.height

# --- Main program to demonstrate the classes ---
if __name__ == "__main__":
    # Create instances of the derived classes
    circle = Circle(radius=5)
    rectangle = Rectangle(width=10, height=8)

    # Call the area() method on each instance and print the result
    print(f"Area of the circle with radius {circle.radius} is: {circle.area():.2f}")
    print(f"Area of the rectangle with dimensions {rectangle.width}x{rectangle.height} is: {rectangle.area()}")

    # Example of polymorphism: a list of shape objects
    shapes = [Circle(2), Rectangle(4, 5), Circle(7)]

    print("\nCalculating areas for a list of different shapes:")
    for shape in shapes:
        if isinstance(shape, Circle):
            print(f"Area of the circle is: {shape.area():.2f}")
        elif isinstance(shape, Rectangle):
            print(f"Area of the rectangle is: {shape.area()}")
```

```python
# 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.
# Ans.
class BankAccount:
    """
    A class to demonstrate encapsulation by modeling a bank account.
    """

    def __init__(self, initial_balance=0):
        """
        Initializes the BankAccount with an optional initial balance.
        """
        if initial_balance < 0:
            print("Initial balance cannot be negative.")
            self.__balance = 0
        else:
            self.__balance = initial_balance  # The private balance attribute

    def deposit(self, amount):
        """
        Deposits a specified amount into the account.

        Args:
            amount: The amount of money to deposit.
        """
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: Rs.{amount:.2f}. New balance: Rs.{self.__balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        """
        Withdraws a specified amount from the account.

        Args:
            amount: The amount of money to withdraw.
        """
        if amount > 0:
            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrew: Rs.{amount:.2f}. New balance: Rs.{self.__balance:.2f}")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    def get_balance(self):
        """
        Returns the current account balance.
        """
        return self.__balance

# --- Demonstration of Encapsulation ---

# 1. Create an instance of the BankAccount class
my_account = BankAccount(1000)

# 2. Use the public methods to interact with the object's data
print(f"Initial balance: Rs.{my_account.get_balance():.2f}")
my_account.deposit(500)
my_account.withdraw(200)

# 3. Attempt to bypass encapsulation and access the private attribute directly
# This will result in an AttributeError
print("\nAttempting to access private balance directly:")
try:
    print(my_account.__balance)
except AttributeError as e:
    print(e)

# 4. Attempt to create an account with a negative balance
print("\nAttempting to create an account with a negative initial balance:")
bad_account = BankAccount(-100)
```

```python
# 7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.
# Ans.
class MathOperations:
    """
    A class to demonstrate class and static methods for mathematical operations.
    """

    @classmethod
    def add_numbers(cls, x, y):
        """
        Adds two numbers using a class method.
        This method is bound to the class and receives the class object 'cls'.
        """
        return x + y

    @staticmethod
    def subtract_numbers(x, y):
        """
        Subtracts two numbers using a static method.
        This method has no access to the class or its instances.
        """
        return x - y

# --- Using the methods ---

# Call the class method directly on the class
sum_result = MathOperations.add_numbers(10, 5)
print(f"The sum of 10 and 5 is: {sum_result}")

# Call the static method directly on the class
difference_result = MathOperations.subtract_numbers(10, 5)
print(f"The difference of 10 and 5 is: {difference_result}")

# You can also call the methods from an instance of the class
math_obj = MathOperations()
sum_from_instance = math_obj.add_numbers(20, 8)
difference_from_instance = math_obj.subtract_numbers(20, 8)
print(f"The sum from an instance 20 & 8 is: {sum_from_instance}")
print(f"The difference from an instance 20 & 8 is: {difference_from_instance}")
```

```python
# 10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.
# Ans.
class Vector:
    """Represents a vector in 2D space."""

    def __init__(self, x=0, y=0):
        """Initializes a new Vector object."""
        self.x = x
        self.y = y

    def __add__(self, other):
        """
        Overloads the '+' operator to add two Vector objects.
        This method is invoked when `vector1 + vector2` is called.
        """
        if isinstance(other, Vector):
            # Adds the corresponding components of the two vectors
            new_x = self.x + other.x
            new_y = self.y + other.y
            return Vector(new_x, new_y)
        else:
            raise TypeError("Unsupported operand type. Can only add another Vector.")

    def __str__(self):
        """
        Overloads the built-in str() function for a human-readable output.
        This allows `print(my_vector)` to output a clean string.
        """
        return f"Vector(x={self.x}, y={self.y})"

# --- Example usage ---

# Create two Vector objects
v1 = Vector(3, 4)
v2 = Vector(1, 2)

# Add the two vectors using the overloaded '+' operator
v3 = v1 + v2

# Print the vectors to see the result
print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Resultant Vector (v1 + v2): {v3}")
```
